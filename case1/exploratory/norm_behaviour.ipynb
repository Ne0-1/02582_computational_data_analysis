```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
def setup_mpl():
    mpl.rcParams['font.family'] = 'Times New Roman'
    return
setup_mpl()
```

```python
#The following visualization is inspired by figure 9.1 page 430 from Machine Learning a Bayesian and Optimization Perspective
theta_vals = np.linspace(-1.5, 1.5, 100)
fig, ax = plt.subplots(figsize=(5,5), dpi=100)
colors = ['firebrick', 'darkblue']
for i, p in enumerate([1, 2]):
    norm_vals = np.power(abs(theta_vals), p)
    ax.plot(theta_vals, norm_vals, '--', c= colors[i], label=f'p={p}')
ax.legend()
ax.set_xlabel(r'$\theta$')
ax.set_ylabel(r'$|\theta|^{p}$', rotation=0)
plt.savefig('norm_behaviour.png', dpi=100)
plt.show()
```
